```python
xstart = 0
xend = 2 * np.pi
increments = 0.1

x = np.arange(xstart, xend, increments)
y = np.sin(x)

# subplot
plt.subplot(2, 1, 1)

plt.title("sin(x)")
plt.grid()
plt.plot(x, y, 'g') # g = color 




# subplot
plt.subplot(2, 1, 2)

z = np.cos(x)

plt.title("sin(x)")
plt.grid()
plt.plot(z, y, 'r') # r = color 


plt.subplots_adjust(wspace=10)

plt.show()

```
